```python
import matplotlib.pyplot as plt
import numpy as np

# 折れ線グラフデータを作成
y_line = [1, 2, -5, 2]

# データを作成 (サイン波)
# 0から10までの幅で配列を等間隔に100個作成する
x_sin = np.linspace(0, 10, 100)
y_sin = 2 + 2 * np.sin(2 * x_sin)

# プロット
# nrowsとncolsを定義
nrows = 2
ncols = 3
fig, axes = plt.subplots(nrows, ncols, figsize=(10, 7))

# 最初のサブプロット (axes[0, 0]) に折れ線グラフを描画
axes[0, 0].plot(y_line)
axes[0, 0].set_title('Line Plot (y_line)')

# 2番目のサブプロット (axes[0, 1]) にサイン波を描画
axes[0, 1].plot(x_sin, y_sin, linewidth=2.0)
axes[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 11),
               ylim=(-1, 5), yticks=np.arange(-1, 6))
axes[0, 1].set_title('Sine Wave Plot')

# 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
x_hist = np.random.normal(15, 5, 2000)
# 3番目のサブプロット (axes[0, 2]) にヒストグラムをプロット
axes[0, 2].hist(x_hist, bins=30)
axes[0, 2].set_title('Histogram')

# データを生成する (散布図用)
np.random.seed(3)
x_scatter = 4 + np.random.normal(0, 2, 60)
y_scatter = 4 + np.random.normal(0, 2, len(x_scatter))

# 4番目のサブプロット (axes[1, 0]) に散布図をプロット
axes[1, 0].scatter(x_scatter, y_scatter)
axes[1, 0].set_title('Scatter Plot')

plt.tight_layout() # サブプロット間のスペースを調整
plt.show()
```
